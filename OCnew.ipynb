```python
from sklearn import set_config
from sksurv.linear_model import CoxPHSurvivalAnalysis
```

```python
data_x1 = merged_oc.drop(columns=['Length FU', 'Outcome'])
data_y1 = merged_oc[['Length FU', 'Outcome']]
merged_oc['Outcome'] = merged_oc['Outcome'].astype(bool)
merged_oc['Length FU'] = merged_oc['Length FU'].astype(int)
data_y1_structured = np.array(list(zip(merged_oc['Outcome'], merged_oc['Length FU'])), dtype=[('Outcome', bool), ('Length FU',int)])
```

```python
model_a = CoxPHSurvivalAnalysis(alpha=1.0)
model_a.fit(data_x1, data_y1_structured)
model_a_predict = model_a.predict(data_x1)
print(model_a_predict)
```

```python
from sksurv.metrics import concordance_index_censored
prediction = model_a.predict(data_x1)
result = concordance_index_censored(data_y1['Outcome'].astype(bool), data_y1['Length FU'], prediction)
print("Concordance Index:", result[0])
```

```python
coefficients_a = pd.Series(model_a.coef_, index=data_x1.columns)
print(coefficients_a)
```

```python
coefficients_a.sort_values(ascending=False).plot.bar()
plt.title('Cox PH Model Coefficients of OC')
plt.show()
```

```python
from sksurv.nonparametric import kaplan_meier_estimator
time = np.array(list(merged_oc['Length FU'].astype(int)))
events = np.array(list(merged_oc['Outcome'].astype(bool)))
survival_probabilities, time_points, conf_int = kaplan_meier_estimator(events, time, conf_type="log-log")
for time_point, survival_probability in zip(time_points, survival_probabilities):
  print(f"Time: {time_point}, Survival Probability: {survival_probability}, Confidence Interval: {conf_int}")
```

```python
%matplotlib inline
plt.step(time_points, survival_probabilities, where="post")
plt.fill_between(time_point, conf_int[0], conf_int[1], alpha=1.0, step="post")
plt.ylabel(r"Estimated probability of survival $\hat{S}(t)$")
plt.xlabel("Time $t$")
```
